# UN DATA ANALYSIS

Analysis on US Data comparing gdp per capita and internet usage

### Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Formatting

```python
%matplotlib inline
```

### Load Data

```python
gdp_df = pd.read_csv('../data/gdp_per_capita.csv')
```

```python
internet_df = pd.read_csv('../data/internet_use.csv', nrows=4496)
```

### DF info

```python
gdp_df.info()
# 6731 rows, 4 columns
# country object, year object, value float64, value footnotes float64
```

```python
internet_df.info()
# 4496 rows, 4 columns
#country or Area object, Year object, value float64, value or footnotes float64
```

### Examine footers

```python
gdp_df[-10:]
```

```python
internet_df[-10:]
```

### Delete Value footnotes

```python
del gdp_df['Value Footnotes']
```

```python
del internet_df['Value Footnotes']
```

### Change Column Names

```python
gdp_df.columns=['Country','Year','Gdp_Per_Capita']
```

```python
internet_df.columns=['Country','Year','Internet_users_Pct']
```

### Merge DFs

```python
gdp_and_internet_use = pd.merge(gdp_df,
                                internet_df,
                                how = 'left',
                                left_on = ['Country', 'Year'],
                                right_on = ['Country', 'Year'])
```
